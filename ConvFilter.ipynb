```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np
```

```
image = mpimg.imread('taj_mahal.jpeg')
plt.imshow(image)
```

```
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

```
sobel_y = np.array([[-1,-2,-1],
                    [0,0,0],
                    [1,2,1]])

filter_image1 = cv2.filter2D(gray, -1, sobel_y)

plt.imshow(filter_image1, cmap = 'gray')
```

```
sobel_x = np.array([[-1,0,1]
                    ,[-2,0,2],
                    [-1,0,1]])
filter_image2 = cv2.filter2D(gray, -1, sobel_x)

plt.imshow(filter_image2, cmap = 'gray')
```

```
myfilter = np.array([[0,-1,0],
                     [-1,4,-1],
                     [0,-1,0]])

filter_image3 = cv2.filter2D(gray, -1, myfilter)

plt.imshow(filter_image3, cmap = 'gray')
```

```
myfilter = np.array([[0,-2,0],
                     [-2,8,-2],
                     [0,-2,0]])

filter_image3 = cv2.filter2D(gray, -1, myfilter)

plt.imshow(filter_image3, cmap = 'gray')
```

```
myfilter = np.array([[0,-1.5,0],
                     [-1.5,6,-1.5],
                     [0,-1.5,0]])

filter_image3 = cv2.filter2D(gray, -1, myfilter)

plt.imshow(filter_image3, cmap = 'gray')
```
